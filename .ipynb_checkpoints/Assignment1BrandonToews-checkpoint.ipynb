```python
import pandas as pd
import numpy as np

# Save data in Pandas dataframe
dataset = pd.read_csv("heart_2020.csv")

# Print how many rows and columns are in dataset
print(dataset.shape)

# Display 1st five entries of dataset
dataset.head()
```

```python
# Print data types and how many null values are present
dataset.info()
```

```python
# Convert object data types to category types
dataset['HeartDisease'] = dataset['HeartDisease'].astype('category')
dataset['Smoking'] = dataset['Smoking'].astype('category')
dataset['AlcoholDrinking'] = dataset['AlcoholDrinking'].astype('category')
dataset['Stroke'] = dataset['Stroke'].astype('category')
dataset['DiffWalking'] = dataset['DiffWalking'].astype('category')
dataset['Sex'] = dataset['Sex'].astype('category')
dataset['Race'] = dataset['Race'].astype('category')
dataset['Diabetic'] = dataset['Diabetic'].astype('category')
dataset['PhysicalActivity'] = dataset['PhysicalActivity'].astype('category')
dataset['GenHealth'] = dataset['GenHealth'].astype('category')
dataset['Asthma'] = dataset['Asthma'].astype('category')
dataset['KidneyDisease'] = dataset['KidneyDisease'].astype('category')
dataset['SkinCancer'] = dataset['SkinCancer'].astype('category')

# Convert float data types to integer types
dataset['PhysicalHealth'] = dataset['PhysicalHealth'].apply(int)
dataset['MentalHealth'] = dataset['MentalHealth'].apply(int)
dataset['SleepTime'] = dataset['SleepTime'].apply(int)

# Replace age ranges with a median number
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['18-24'], 21)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['25-29'], 27)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['30-34'], 32)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['35-39'], 37)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['40-44'], 42)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['45-49'], 47)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['50-54'], 52)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['55-59'], 57)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['60-64'], 62)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['65-69'], 67)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['70-74'], 72)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['75-79'], 77)
dataset['AgeCategory'] = dataset['AgeCategory'].replace(['80 or older'], 82)

# Display 1st five entries of dataset showing cleaned data
dataset.head()
```

```python
# Print new data types and how many null values are present
dataset.info()
```
